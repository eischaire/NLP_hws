Уберём ненужные нам вещи типа примечаний, юникод-символов и т д

```python
def ignore_unicode(text):
    text = re.sub('\\xa0', ' ', text)
    text = re.sub('(\[.+?\])', '', text)
    text = re.sub('\n\n', '\n', text)
    return text
```

Обучим токенайзер

```python
with open('tourgueniev_2.txt', 'r', encoding='utf-8') as k:
    data = k.read()
    data = ignore_unicode(data)
```

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(data)
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

```python
with open('tourgueniev_1_plain.txt', 'r', encoding='utf-8') as n:
    test = n.read()
    test = ignore_unicode(test)
```

```python
tokenizer.tokenize(test)
```

Посчитаем метрики

```python
def metrics(sent_tokenize, gold):
    tp = 0
    fp = 0
    fn = 0

    for sent in gold:
        if len(sent_tokenize(sent)) == 1:
            tp += 1
        else:
            fp += 1

    for i in range(len(gold)-1):
        sent1, sent2 = gold[i], gold[i+1]
        sent = ' '.join([sent1, sent2])
        if len(sent_tokenize(sent)) == 2:
            tp += 1
        else:
            fn += 1

    precision = (tp/(tp+fp))
    recall = (tp/(tp+fn))
    f1 = 2*(precision*recall)/(precision+recall)
    print('Precision - ', precision)
    print('Recall - ', recall)
    print('F1 - ', f1)
```

```python
metrics(tokenizer.tokenize, ideal)
```
